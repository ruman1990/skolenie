```python
import matplotlib.pyplot as plt
import numpy as np

# Graf 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x, y)

# Graf 2
y2 = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x, y2)

plt.show()
```

```python
plt.scatter(x,y2)
plt.show()
```

```python
plt.bar(x,y2)
plt.xlabel('Hodnoty')
plt.ylabel('Y hodnoty')
plt.title('Super graf')
plt.show()
```
